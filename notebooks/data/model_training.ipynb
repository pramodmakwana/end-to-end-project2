```python
# Define the custom ranking for each ordinal variable
cut_categories={"Fair":1,"Good":2,"Very Good":3,"Premium":4,"Ideal":5}
clarity_categories={"I1":1,"SI2":2,"SI1":3,"VS2":4,"VS1":5,"VVS2":6,"VVs1":7,"IF":8}
color_categories={"D":1,"E":2,"F":3,"G":4,"H":5,"I":6,"J":7}
```

```python
from sklearn.impute import SimpleImputer ##Handling Missing Values
from sklearn.preprocessing import StandardScaler  ## handling Feturescaling
from sklearn.preprocessing import OrdinalEncoder # Ordinal Encoding

## pipelines
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
```

```python
num_pipeline=Pipeline(
    
    steps=[
        
        ('imputer',SimpleImputer()),
        ('scaler',StandardScaler())
    ]
)
```

```python
cat_pipeline=Pipeline(
    
    steps=[
        
        ('imputer',SimpleImputer(strategy='most_frequent')),
        ('ordinalencoder',OrdinalEncoder(categories=[cut_categories,color_categories,clarity_categories])),
        ('scaler',StandardScaler())
    ]
    
)
```

```python
preprocessor=ColumnTransformer(
    
    [
        ('num_pipeline',num_pipeline,numerical_cols),
        ('cat_pipeline',cat_pipeline,categorical_cols)
    
    ]
)
```

```python
## train test split

from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.30,random_state=30)
```
